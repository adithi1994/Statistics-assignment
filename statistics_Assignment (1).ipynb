```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
data = { "Group": ["A", "A", "A", "A", "A", "B", "B", "B", "B", "B"], "Scores": [70, 65, 75, 80, 68, 85, 88, 84, 90, 87] } 
df= pd.DataFrame(data)
data
```

```python
Group_A_Score = df[df['Group'] == 'A']['Scores']
Group_B_Score = df[df['Group'] == 'B']['Scores']
```

```python
# Conduct Independent Two-Sample T-Test
t_stat, p_value = ttest_ind(Group_A_Score, Group_B_Score, equal_var=False)

# Display results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# Set significance level (alpha)
alpha = 0.05

if p_value < alpha:
    print("Result: Reject the Null Hypothesis (H₀)")
    print("Conclusion: There is a significant difference between the two teaching methods.")
else:
    print("Result: Fail to Reject the Null Hypothesis (H₀)")
    print("Conclusion: No significant difference between the two teaching methods.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a boxplot for score distribution
plt.figure(figsize=(6, 4))
sns.boxplot(x="Group", y="Scores", data=df, palette=["red", "blue"])
plt.title("Comparison of Scores: Traditional vs. New Teaching Method")
plt.xlabel("Teaching Method")
plt.ylabel("Scores")
plt.show()
```

```python
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
group_a = [15, 18, 16, 14, 17] # Plant growth from Fertilizer A 

group_b = [22, 25, 24, 23, 26] # Plant growth from Fertilizer B  

group_c = [30, 35, 32, 33, 34] # Plant growth from Fertilizer C 
```

```python
f_stat,p_value = f_oneway(group_a,group_b,group_c)

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
```

```python
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Result: Reject the Null Hypothesis (H₀)")
    print("Conclusion: There is a significant difference in plant growth due to the fertilizer type.")
else:
    print("Result: Fail to Reject the Null Hypothesis (H₀)")
    print("Conclusion: No significant difference in plant growth due to the fertilizer type.")
```

```python
#for visualization
df= pd.DataFrame({
    "Growth": group_a + group_b + group_c,
    "Fertilizer": ["A"]*5 + ["B"]*5 +["C"]*5
})
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x="Fertilizer", y="Growth", data=df, estimator=sum, palette="pastel")
plt.title("Total Plant Growth for Each Fertilizer")
plt.xlabel("Fertilizer Type")
plt.ylabel("Total Plant Growth (cm)")
plt.show()
```
